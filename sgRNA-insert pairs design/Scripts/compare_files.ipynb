```python
import pandas as pd

# Load both CSV files
df1 = pd.read_csv('../Example_Data/sgRNA_insert_list.csv')
df2 = pd.read_csv('../Example_Data/example_sgRNA_inserts_2.csv')


# Define the columns you want to compare
columns_to_compare = [
    "gene", "parent aa", "parent codon", "aa position", "mutated aa",
    "child codon", "nt position", "dist mut pam", "pam"
]

# Subset the dataframes to only the relevant columns
df1_subset = df1[columns_to_compare].copy()
df2_subset = df2[columns_to_compare].copy()

# Remove duplicate rows if needed (optional)
df1_unique = df1_subset.drop_duplicates()
df2_unique = df2_subset.drop_duplicates()

# Find rows unique to df1
unique_to_df1 = pd.merge(df1_unique, df2_unique, how='outer', indicator=True)\
                  .query('_merge == "left_only"')\
                  .drop(columns=['_merge'])

# Find rows unique to df2
unique_to_df2 = pd.merge(df2_unique, df1_unique, how='outer', indicator=True)\
                  .query('_merge == "left_only"')\
                  .drop(columns=['_merge'])

# Print results
print("\nRows unique to File 1:")
print(unique_to_df1.to_string(index=False))

print("\nRows unique to File 2:")
print(unique_to_df2.to_string(index=False))
```
