```python
import pandas as pd

# Load both CSV files
df1 = pd.read_csv('../Example_Data/sgRNA_insert_list.csv')
df2 = pd.read_csv('../Example_Data/example_sgRNA_inserts.csv')


# Concatenate and find symmetric difference
combined = pd.concat([df1, df2]).drop_duplicates(keep=False)

# Show result
print(f"Number of unique rows only in one file: {len(combined)}")
print(combined)
```

```python
# Ensure the columns are in the same order and have the same names
df1 = df1.sort_index(axis=1)
df2 = df2.sort_index(axis=1)

# Merge with indicator to find which rows are unique to each
merged = pd.merge(df1, df2, how='outer', indicator=True)

# Filter rows that are only in one of the two dataframes
diff_rows = merged[merged['_merge'] != 'both'].drop(columns=['_merge'])

print(f"Number of rows only in one file: {len(diff_rows)}")
print(diff_rows)
```
