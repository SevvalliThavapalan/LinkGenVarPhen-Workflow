```python
import pandas as pd

# Load both CSV files
df1 = pd.read_csv('../Example_Data/sgRNA_insert_list.csv')
df2 = pd.read_csv('../Example_Data/example_sgRNA_inserts.csv')


# Concatenate and find symmetric difference
combined = pd.concat([df1, df2]).drop_duplicates(keep=False)

# Show result
print(f"Number of unique rows only in one file: {len(combined)}")
print(combined)
```
